```python
import json

traj_path = "exports/tau_bench_retail_20250917_062413/dumped_evals/global_step_30_evals/test_retail/retail.jsonl"
```

```python
trajectories = []
with open(traj_path, 'r') as f:
    for line in f:
        traj = json.loads(line)
        trajectories.append(traj)
        
print(f"Number of trajectories: {len(trajectories)}")

print(f"Fields in the first trajectory: {trajectories[0].keys()}")
```

```python
extra_info = json.loads(trajectories[0]['env_extras']['extra_info'])
print(extra_info.keys())
print(extra_info['actions'])
```

```python
index = 10
print(trajectories[index]['env_extras'].keys())
reward_spec = json.loads(trajectories[index]['env_extras']['reward_spec'])
print(reward_spec.keys())
print(f"Ground truth actions:")
for action in reward_spec['ground_truth']:
    print(action)
```

```python
print(trajectories[index]['output_response'])
```

```python
for message in trajectories[index]['conversation']:
    print("Role:", message['role'])
    print("Content:", message['content'])
    print("-" * 100)
```
