```python
import json
# /Users/mertcemri/Desktop/initials/data/reward_filtered_sft_data_all_domains.jsonl
data = []
with open('../../data/reward_filtered_sft_data_all_domains.jsonl', 'r') as f:
    for line in f:
        if line.strip():  # Skip empty lines
            data.append(json.loads(line))

print(f"Loaded {len(data)} records")
print("First record:", data[0] if data else "No data")
```
